```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
import pandas as pd
import numpy as np
np.random.seed(2018)
```

```python
train_data = pd.read_csv('CDM-RN-2/train.csv')

train_data['Sex']=np.where(train_data['Sex']=='female',0,1)
del train_data['PassengerId']
del train_data['Name']
del train_data['Ticket']
del train_data['Fare']
del train_data['Cabin']
del train_data['Embarked']
dataset = train_data.dropna()
dataset.reindex(np.random.permutation(dataset.index))
```

### 1.- Dataset

```python
x_train = dataset.iloc[:614,[1,2,3,4,5]]
```

```python
x_test = dataset.iloc[614:714,[1,2,3,4,5]]
```

```python
y_train = dataset.iloc[:614,[0]]
```

```python
y_test = dataset.iloc[614:714,[0]]
```

### 2.- Estructura de la RN

```python
model = Sequential()
```

```python
model.add(Dense(___, input_dim=5, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(___, activation='sigmoid'))

model.compile(loss='binary_crossentropy', 
              optimizer='adam',
              metrics=['accuracy'])
```

### 3.- Entrenamiento y testeo

```python
model.fit(x_train, y_train, epochs=___, batch_size=8)
```

```python
evaluation = model.evaluate(x_test, y_test)
```
